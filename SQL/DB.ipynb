```python
import pymysql
import pandas as pd
import configparser
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('../data/clean_tweets.csv')
df.head()
```

```python
config = configparser.ConfigParser()
config.read('config.ini')
```

```python
db = pymysql.connect(host = config['RDS']['host'],
                     user = config['RDS']['user'],
                     password = config['RDS']['password'],
                     cursorclass = pymysql.cursors.DictCursor
)

cursor = db.cursor()
```

CREATE DATABASE

```python
create_db = '''CREATE DATABASE twitter_database'''
# cursor.execute(create_db)
```

```python
cursor.execute('SHOW DATABASES')
cursor.fetchall()
```

# Create tables

```python
cursor.connection.commit()
use_db = '''USE twitter_database'''
cursor.execute(use_db)
```

```python
create_table = '''
CREATE TABLE tweet (
id INT,
tweet TEXT,
created_at DATE
)
'''

cursor.execute(create_table)

create_table = '''
CREATE TABLE user (
id INT,
user_id INT,
username TEXT,
retweets_count INT,
replies_count INT,
likes_count INT,
quote_url TEXT
)
'''

cursor.execute(create_table)
```

```python
# drop_table = '''DROP TABLE tweet'''
# cursor.execute(drop_table)
```

```python
cursor.execute('SHOW TABLES')
cursor.fetchall()
```

# SQL

```python
engine = create_engine(
    "mysql+pymysql://{user}:{pw}@{host}/{db}".format(
        user = config['RDS']['user'],
        pw = config['RDS']['password'],
        host = config['RDS']['host'],
        db = 'twitter_database'))
```

```python
df[['id', 'tweet', 'created_at']].to_sql(
    name='tweet',
    con=engine,
    if_exists= 'append',
    index=False)

df[['id', 'user_id', 'username', 'retweets_count', 'replies_count', 'likes_count', 'quote_url']].to_sql(
    name='user',
    con=engine,
    if_exists= 'append',
    index=False)
```

```python
db.commit()
```

```python
sql = '''SELECT * FROM tweet'''
cursor.execute(sql)
tweets = cursor.fetchall()
tweets
```

```python
sql = '''SELECT * FROM user'''
cursor.execute(sql)
users = cursor.fetchall()
users
```
